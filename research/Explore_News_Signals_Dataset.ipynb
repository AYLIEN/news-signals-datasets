# Exploring a News Signals dataset

We can use `print` on the dataset to get some information about it.

```python
print(dataset)
```

A signals dataset contains multiple `Signal` objects. In this dataset, each of these signals corresponds to one company of the first 10 entries in the Nasdaq-100 index.

```python
signals = sorted(dataset.signals.values())
[s.name for s in signals]
```

## Exploring signals of a particular entity

Let's pick one of the signals (corresponding to one company/entity) and explore it's data.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [20, 6]
```

```python
signal = signals[0]
```

### Timeseries and Feeds

Each signal has two main collections of sub-signals which are all ordered along the same datetime index and stored as two Pandas dataframes:
* `signal.timeseries_df` stores numerical timeseries, e.g. news publication volume
* `signal.feeds_df` stores textual feeds, in this case news articles about the company published at each day of the time span

```python
signal.timeseries_df.tail(3)
```

```python
signal.feeds_df.tail(3)
```

These are also available combined, simply as `signal.df`:

```python
signal.df.tail(3)
```

We can plot the signal's time series.

```python
signal.plot()
```

```python
signal["count"].plot()
```

```python
signal["anomalies"].plot()
```

Let's see which day has the highest publication count for this entity and let's look at a sample of news articles about the entity published on that day.

```python
max_count_day = signal.df.loc[signal["count"].idxmax()]
max_count_day
```

```python
for s in max_count_day.stories:
    print(s["title"])
```

```python
max_count_day.stories[0]["body"]
```
